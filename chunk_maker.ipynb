```python
import os
import json

base_dir = "/content/sec-edgar-filings"
output_json = "/content/processed_filings.json"

filings_data = []

for root, dirs, files in os.walk(base_dir):
    for file in files:
        if file == "full-submission.txt":
            full_path = os.path.join(root, file)

            # Extract metadata from path: ticker, filing type, filing id
            parts = full_path.split(os.sep)
            # Example path parts: ['', 'content', 'sec-edgar-filings', 'AAPL', '10-K', '0001193125-09-214859', 'full-submission.txt']
            ticker = parts[3]
            filing_type = parts[4]
            filing_id = parts[5]

            # Read the filing text
            with open(full_path, "r", encoding="utf-8", errors="ignore") as f:
                text = f.read()

            filings_data.append({
                "ticker": ticker,
                "filing_type": filing_type,
                "filing_id": filing_id,
                "file_path": full_path,
                "text": text
            })

# Save to JSON file
with open(output_json, "w", encoding="utf-8") as f_out:
    json.dump(filings_data, f_out)

print(f"Processed {len(filings_data)} filings saved to {output_json}")
```

```python
import re

def chunk_text(text, max_tokens=500):
    # Split into sentences
    sentences = re.split(r'(?<=[.?!])\s+', text)
    chunks = []
    current_chunk = []

    current_length = 0
    for sentence in sentences:
        length = len(sentence.split())
        if current_length + length > max_tokens:
            chunks.append(" ".join(current_chunk))
            current_chunk = [sentence]
            current_length = length
        else:
            current_chunk.append(sentence)
            current_length += length

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

# Load the existing filings
with open("/content/processed_filings.json", "r") as f:
    filings = json.load(f)

# Apply chunking
chunked_data = []
for filing in filings:
    chunks = chunk_text(filing["text"])
    for idx, chunk in enumerate(chunks):
        chunked_data.append({
            "ticker": filing["ticker"],
            "filing_type": filing["filing_type"],
            "filing_id": filing["filing_id"],
            "chunk_id": idx,
            "chunk_text": chunk
        })

# Save chunked output
with open("/content/chunked_filings.json", "w") as f:
    json.dump(chunked_data, f)

print(f"Total chunks created: {len(chunked_data)}")
```

```python
# —— User upload section —— #
print("► If you have your own doc (txt, pdf, docx, html), upload it here:")
uploaded = colab_files.upload()  # pops up the browse dialog

for fname, filebytes in uploaded.items():
    ext = fname.split('.')[-1].lower()
    if ext == 'txt' or ext == 'html':
        text = filebytes.decode('utf-8', errors='ignore')
    elif ext == 'pdf':
        reader = PdfReader(io.BytesIO(filebytes))
        pages = [p.extract_text() for p in reader.pages]
        text = "\n".join(pages)
    elif ext == 'docx':
        doc = Document(io.BytesIO(filebytes))
        text = "\n".join(p.text for p in doc.paragraphs)
    else:
        print(f"Skipping unsupported file type: {fname}")
        continue

    # reuse your chunk_text
    new_chunks = chunk_text(text)
    base_meta = {
        "ticker": "USER_UPLOAD",
        "filing_type": ext,
        "filing_id": fname
    }
    for idx, chunk in enumerate(new_chunks):
        chunked_data.append({
            **base_meta,
            "chunk_id": idx,
            "chunk_text": chunk
        })

print(f"After upload, total chunks = {len(chunked_data)}")
# Save the augmented chunks
with open("/content/chunked_filings.json", "w") as fout:
    json.dump(chunked_data, fout)
print("✅ Saved updated /content/chunked_filings.json")
```

```python
import json
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer
from typing import List, Tuple

class FinancialRetriever:
    def __init__(self, model_name: str = 'all-MiniLM-L6-v2'):
        self.model = SentenceTransformer(model_name)
        self.chunk_data = []
        self.texts = []
        self.index = None

    def load_chunks(self, path: str):
        """Load and parse JSON chunks from the given path"""
        with open(path, "r") as f:
            self.chunk_data = json.load(f)
        self.texts = [entry["chunk_text"] for entry in self.chunk_data]
        print(f"Loaded {len(self.texts)} chunks.")

    def build_index(self):
        """Generate embeddings and build FAISS index (normalized, cosine similarity)"""
        print("Generating embeddings...")
        embeddings = self.model.encode(
            self.texts, batch_size=64, show_progress_bar=True
        )
        matrix = np.array(embeddings).astype('float32')
        faiss.normalize_L2(matrix)
        self.index = faiss.IndexFlatIP(matrix.shape[1])
        self.index.add(matrix)
        print(f"FAISS index built with {self.index.ntotal} vectors.")

    def embed_query(self, query: str) -> np.ndarray:
        vec = self.model.encode([query], convert_to_numpy=True)
        faiss.normalize_L2(vec)
        return vec.astype('float32')

    def search(self, query: str, top_k: int = 5) -> List[Tuple[str, float, dict]]:
        if self.index is None:
            raise ValueError("Index not built.")
        qv = self.embed_query(query)
        distances, indices = self.index.search(qv, top_k)
        results = []
        for rank, idx in enumerate(indices[0]):
            text = self.texts[idx]
            meta = self.chunk_data[idx]
            score = float(distances[0][rank])
            results.append((text, score, meta))
        return results

    def format_results(self, results):
        return [
            {
                "preview": text[:300] + "...",
                "score": round(score, 4),
                "source": meta.get("source_file", "unknown")
            }
            for text, score, meta in results
        ]
```

```python
json_path = 'chunked_filings.json'

retriever = FinancialRetriever('all-MiniLM-L6-v2')
retriever.load_chunks(json_path)
retriever.build_index()

# 5. Sample query
query = "Apple's financial performance in 2023"
results = retriever.search(query, top_k=5)
formatted = retriever.format_results(results)

for i, r in enumerate(formatted, 1):
    print(f"Rank {i}: score={r['score']}, source={r['source']}")
    print(r['preview'])
    print('-'*80)
```
